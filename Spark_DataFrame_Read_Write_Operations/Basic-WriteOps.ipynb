#Datalake
We are writing data in Structured(csv), Semi Structured(JSON/XML), Serialized files (orc/parquet/delta) <br/>


#lakehouse
Table (delta/hive) 

```python


from pyspark.sql.session import SparkSession
spark=SparkSession.builder.appName("Spark DataFrames").getOrCreate()
```

#Extracting the data

```python
ingest_df1=spark.read.csv("/Volumes/workspace/ingestion_volume/source/custs_header",header=True,sep=',',inferSchema=True,samplingRatio=0.10)
```

#Loading the data

```python
ingest_df1.write.csv(path="/Volumes/workspace/ingestion_volume/target/csvout",sep='~',header=True,mode='overwrite')
spark.read.csv("/Volumes/workspace/ingestion_volume/target/csvout").show(2)
     
```

#Transforming the data

```python
transformed_df=ingest_df1.select("custid","fname","lname","profession","age").withColumnRenamed("profession","prof")

     
```

```python
transformed_df.write.csv(path="/Volumes/workspace/ingestion_volume/target/csvout",sep='~',header=True,mode='overwrite')
#compressed data
transformed_df.write.csv(path="/Volumes/workspace/ingestion_volume/target/csvout",sep='~',header=True,mode='overwrite',compression='gzip')
```

#Json format

```python
ingest_df1.write.json(path="/Volumes/workspace/ingestion_volume/target/jsonout",mode='overwrite')
spark.read.json("/Volumes/workspace/ingestion_volume/target/jsonout").show(2)
```

#Serialization (encoding in a more optimized fashion) <br/>Deserialization File formats (Binary/Brainy File formats)


```python

ingest_df1.write.orc(path="/Volumes/workspace/ingestion_volume/target/orcout",mode='overwrite',compression='zlib')#by default orc/parquet uses snappy compression
spark.read.orc("/Volumes/workspace/ingestion_volume/target/orcout").show(2)#uncompression + deserialization
```

#Orc/Parquet follows WORM feature (Write Once Read Many)

```python

ingest_df1.write.mode("overwrite").option("compression","snappy").parquet(path="/Volumes/workspace/ingestion_volume/target/parquetout")#by default orc/parquet uses snappy compression
spark.read.parquet("/Volumes/workspace/ingestion_volume/target/parquetout").show(2)#uncompression + deserialization
     
```

#Delta follows WMRM feature (Write Many Read Many)

```python

ingest_df1.write.format("delta").save("/Volumes/workspace/ingestion_volume/target/deltaout",mode='overwrite')
spark.read.format("delta").load("/Volumes/workspace/ingestion_volume/target/deltaout").show(2)
```
